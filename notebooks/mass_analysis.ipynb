```python
def make_sentence_positions(tensor, padding_idx, eos_idx):
    mask = tensor.ne(padding_idx).int()
    return (
        torch.cumsum((tensor.type_as(mask) * mask).eq(eos_idx), dim=0).type_as(mask) + mask 
    ).long() + padding_idx


```
