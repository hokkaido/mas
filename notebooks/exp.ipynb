```python
giga_path = '../datasets/sumdata/train'
```

```python
import os
import random

def sample_giga(k=50):
    fa = os.path.join(giga_path, 'train.article.txt')
    ft = os.path.join(giga_path, 'train.title.txt')
    examples = []
    with open(fa) as articles:
        with open(ft) as titles:
            for n, article in enumerate(articles):
                title = titles.readline()
                example = (article.strip().split(), title.strip().split())
                if n < k:
                     examples.append(example)
                else:
                    r = random.randint(0, n)
                    if r < k:
                        examples[r] = example
    return examples
```

```python
examples = sample_giga(25)
```

```python
examples
```

```python
from collections import Counter

def shared_tokens(a, b):
    ac = Counter(a)
    bc = Counter(b)
    return (ac & bc).keys()
```

```python
for ex in examples:
    print(shared_tokens(*ex))
```
